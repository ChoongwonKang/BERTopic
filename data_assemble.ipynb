```python
import pandas as pd
import os

# 폴더 경로 설정
folder_path = os.getcwd()

# 모든 엑셀 파일 목록 가져오기
file_list = os.listdir(folder_path)
excel_files = [file for file in file_list if file.endswith('.xlsx')]

# 데이터를 저장할 빈 DataFrame 생성
combined_data = pd.DataFrame()

# 각 엑셀 파일 읽어와서 DataFrame에 추가하기
for file in excel_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_excel(file_path)
    
    # 파일명을 값으로 갖는 새로운 열 추가
    df['파일명'] = file
    
    # 데이터를 빈 DataFrame에 추가
    combined_data = combined_data.append(df, ignore_index=True)

# 결과 출력
print(combined_data)
```

```python
combined_data
```

```python
df_combined = combined_data

df = df_combined.drop_duplicates(subset=['뉴스 식별자'])
df
```

```python
#월별 분석을 위한 월 분리 열 생성
from datetime import datetime
def monthly(text) :
    month = str(text)[:6]
    return datetime.strptime(month, '%Y%m')
```

```python
df
```

```python
df['월별']= df['일자'].apply(monthly)

df.head()
```
