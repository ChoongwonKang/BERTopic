```python
#DF(Document Frequency) 계산기
from sklearn.feature_extraction.text import CountVectorizer
def calculate_df(df):
    token_list = df['Token'].map(lambda x: ' '.join(x)).tolist()
    vect = CountVectorizer(tokenizer=str.split)
    document_term_matrix = vect.fit_transform(token_list)
    tf = pd.DataFrame(document_term_matrix.toarray(), columns=vect.get_feature_names_out())  
    DF = tf.astype(bool).sum(axis = 0)
    DF_df= pd.DataFrame(DF, columns=['Frequency']).sort_values(by='Frequency', ascending =False)

    return DF_df   

def calculate_df_topic(df, topic) :
    topic_df = df[df['Topic']==topic]
    result_df=calculate_df(topic_df)
    return result_df

def calculate_df_keyword(df, keyword_list) :
    token_list = df['Token'].map(lambda x: ' '.join(x)).tolist()
    vect = CountVectorizer(tokenizer=str.split)
    document_term_matrix = vect.fit_transform(token_list)
    tf = pd.DataFrame(document_term_matrix.toarray(), columns=vect.get_feature_names_out()) 
    filtered_tf = tf[tf[keyword_list].any(axis=1)]
    return len(filtered_tf) 
```

```python
# 함수로 만들기
import numpy as np

def get_acc(x=[], y=[]) : # 입력값 리스트로 받기(수치 데이터)
    
    x = np.array(x) #넘파이 배열로 변경
    y= np.array(y)
    #print(x)
    #print(y)
    coefficients = np.polyfit(x, y, 2) #2차 회귀곡선 찾기

    acceleration= coefficients[0]#가속도 = x 제곱의 계수
    
    average = y.mean()
    print(f"The average is: {average}")
    print(f"The acceleration is: {acceleration}")

    return acceleration, average
```
